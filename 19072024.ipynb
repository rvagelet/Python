```python
import pandas as pd
Pedidos = pd.read_excel(r"/content/Datos Pedidos.xlsx")
Pedidos
```

```python
import pandas as pd
DFPedidos = Pedidos['Cliente']
DFPedidos = (Pedidos['Pais']=='Italia') | (Pedidos['Pais']=='Alemania')
Pedidos[DFPedidos][['Cliente','Ciudad','Region','Pais']].sort_values('Pais',ascending=True,ignore_index=True)


```

```python
DF2Pedidos = pd.DataFrame(Pedidos, columns=['Cliente','Ciudad','Region','Pais'])
DF2Pedidos.info()
DF2Pedidos
```

```python
DF2Pedidos.to_csv("DF2Pedidos.csv",sep=",",index=False)
```

```python
def descuento(Categoria):
     if Categoria == 'Bebidas' or Categoria == 'Lácteos':
          return 0.05
     else:
          return 0

#Crear la columna que contendrá el descuento
Pedidos['DescuentoPorCategoria'] = Pedidos['Importe']*Pedidos['Categoria'].apply(descuento)
Pedidos

Pedidos['Importe Neto'] = Pedidos['Importe'] - Pedidos['DescuentoPorCategoria']
Pedidos
```

```python
#Importe neto por Pais y Ciudad

#Vamos a almacenar la agrupación por categoria
Importeagrupacio = Pedidos.groupby(['Pais','Ciudad'])['Importe Neto'].sum()
Importeagrupacio


```

```python
#Después queremos un grafico circular con FULL EQUIP (con todo), para ver el porcentaje de cantidad de pedidos por comercial.
Importeagrupacio = Pedidos.groupby(['Comercial'])['Cantidad'].count()
Importeagrupacio
```

```python
DFComercial=pd.DataFrame(Importeagrupacio, columns = ['Cantidad']).reset_index()
DFComercial
```

```python
#Crear gráfico circular
import matplotlib.pyplot as plt     #importació del matplotlib
fig, comercial=plt.subplots()       #això és per fer el quadrat blanc. S'ha de reservar espai en el Kernel perquè creï la figura
#podem agafar la fila que vulguem, perquè totes elles ens permetran sumar les fileres!
comercial.pie(DFComercial['Cantidad'], labels=DFComercial.index, autopct="%.2f%%")
plt.show
```

```python
#Suma importe neto
Importeagrupacio1 = Pedidos.groupby(['Pais'])['Importe Neto'].count()
Importeagrupacio1
```

```python
DFPais=pd.DataFrame(Importeagrupacio1, columns = ['Importe Neto']).reset_index()
DFPais
```

```python
#Queremos un grafico de barras FULL EQUIP, donde ver la suma por importe neto por país
#Grafico de columnas. Llamaremos a matplotlib
import matplotlib.pyplot as plt
#el grafico de columnas seria tipo bar, reservar espacio para la figura
fig, sumapais=plt.subplots()
#primero la dimensión y después la métrica
sumapais.bar(DFPais['Pais'],DFPais['Importe Neto'], color='red')
#con bars podremos hacer configuraciones más
sumapais.set_xlabel('Països')
plt.xticks(rotation=90)    #para poner las etiquetas rotadas en el eje x
sumapais.set_ylabel('Import')
plt.rcParams["figure.figsize"]=(28,6)     #la linea redimensiona la figura (ancho, alto)
plt.show
```
