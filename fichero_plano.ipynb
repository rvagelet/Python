```python
#Practica 5: Ver los ultimos 10 nombres de los pasajeros que son mujer ordenados de manera ascendente
#Practica 6: Ver en un texto el numero de pasajeros que sobrevivieron (survived = 0 No sobrevivió/ 1 sobrevivió)
#Practica 7: Guardar en un nuevo DataFrame solo los 15 primeros tiquets cuyo nombre contenga un 5
```

```python
#Practica 5: Ver los ultimos 10 nombres de los pasajeros que son mujer ordenados de manera ascendente
FTSF = titanic['Sex'] == 'female'
titanic[FTSF]['Name'].tail(10).sort_values(ascending=True)
```

```python
#Practica 6: Ver en un texto el numero de pasajeros que sobrevivieron (survived = 0 No sobrevivió/ 1 sobrevivió)
FTS1 = titanic['Survived']==1
Numero = titanic[FTS1]['PassengerId'].count()
print('El numero de pasajeros que sobrevivieron es', Numero)
```

```python
#Practica 7: Guardar en un nuevo DataFrame solo los 15 primeros tiquets cuyo nombre contenga un 5
FT15T = titanic['Ticket'].str.contains('5')
titanic15T = titanic[FT15T].head(15)
titanic15T
```

```python
#Como aplicar mas de una condicion a un filtro, para ello usaremos los operadores de logic OR + AND
#en el caso del AND le pondremos & (Ampersand) y en el caso del OR este operador logico | (altGr + la tecla del 1)
FTC1C3 = (titanic['Pclass'] == 1) | (titanic['Pclass'] == 3)
titanic[FTC1C3]
```

```python
#Como aplicar el AND, es decir que se cumplan todas las condiciones. Clase 2 y embarcar por la puerta S
FTC2ES = (titanic['Pclass'] == 2)&(titanic['Embarked'] == 'S')
titanic[FTC2ES]
```

```python
#Para ver los diferentes valores que tiene el campo, con unique()
titanic['Embarked'].unique()
```

```python
#Practica 8: Ver los pasajeros que son hombres de la clase 2.
#Practica 9: Ver la cantidad de pasajeros que son hombres de la clase 1 y que embarcaron por la puerta S.
#Practica 10: Ver los nombres de los pasajeros de la clase 3 que no sobrevivieron
#Practica 11: Ver los 12 últimos nombres en orden descendente de las mujeres de la clase 3 cuyo PassengerId es mayor de 100
#Practica 12: Ver en una lista los nombres de los pasajeros cuyo PassengerId está comprendido entre 70 y 180
```

```python
#Practica 8: Ver los pasajeros que son hombres de la clase 2.
FTPHC2 = (titanic['Sex']=='male') & (titanic['Pclass']==2)
titanic[FTPHC2]
```

```python
#Practica 9: Ver la cantidad de pasajeros que son hombres de la clase 1 y que embarcaron por la puerta S.
FTPHC1yPS = ((titanic['Sex']=='male') & (titanic['Pclass']==1)) & (titanic['Embarked']=='S')
titanic[FTPHC1yPS]['Sex'].count()
```

```python
#Practica 10: Ver los nombres de los pasajeros de la clase 3 que no sobrevivieron
FTPC3yNS = (titanic['Pclass']==3) & (titanic['Survived']==0)
titanic[FTPC3yNS]['Name']
```

```python
#Practica 11: Ver los 12 últimos nombres en orden descendente de las mujeres de la clase 3 cuyo PassengerId es mayor de 100
FTMC3Py100 = (titanic['Pclass']==3) & (titanic['PassengerId']>100) & (titanic['Sex'] == 'female')
titanic[FTMC3Py100]['Name'].tail(12).sort_values(ascending=False)
```

```python
#Practica 12: Ver en una lista los nombres de los pasajeros cuyo PassengerId está comprendido entre 70 y 180
FTNPyPI = (titanic['PassengerId']>70) & (titanic['PassengerId']<180)
titanic[FTNPyPI]['Name']
```
